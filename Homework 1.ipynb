# Question one

<h2 style="color:#fc8803 "> Determine the total sales per invoice and identify the top 5 invoices </h2>

### Steps:
- Add column to df named total(quantity * price)
- Group data by invoice number
- Sum all totals for each invoice
- Display the top 5 invoices

```python
# step 1
def calculate_total(quantity, price):
    return quantity * price

df['Total'] = calculate_total(df['Quantity'], df['Price'])
df.head()
```

```python
#step 2:4
top_sales_invoices = df.groupby(['Invoice'])['Total'].sum().sort_values(ascending=False)
top_sales_invoices.head()
```

<h2 style="color:#90fc03;"> Result:</h2>

**Here we can see the top 5 sales in the selected df sample, the invoice '515944' have total of 12366.00$**

# Question 4

<P style="color: #00a6ff; font-family: cursive; font-size: medium"> The retail store aims to introduce a new promotion targeting items with low sales volume. This promotion entails offering a discounted price for a pair of two different items, with one item being among the top 10 best sellers and the other from the bottom 10 sellers. Customers who purchase these two items together will qualify for the special pricing. Your task is to recommend 10 pairs of items to kickstart this campaign. </P>

<h3> Steps: </h3>
<ol>
    <li>Group data by StockCode and description, then sum up the total quantity sold for each item</li>
    <li>Save the the items where the quantity is above 0 to 'top_sold_items' var</li>
    <li>Save the the items where the quantity is equal and below 0 to 'zero_sold_items' var</li>
    <li>Create pairs, one item from the top and the other from the bottom</li>
    </ol>

```python
grouped_items_sold = df.groupby(['StockCode'])['Quantity'].sum().sort_values().to_frame()
grouped_items_sold = grouped_items_sold.rename(columns={'Quantity': 'QuantitySold'})

stock_code_col = grouped_items_sold.index
grouped_items_sold['index'] = [i for i in range(len(grouped_items_sold))]
grouped_items_sold = grouped_items_sold.set_index(['index'])
grouped_items_sold['StockCode'] = stock_code_col

grouped_items_sold
```

<h2 style="color: red;"> Exaplantion:</h2>

- **I grouped items based on their StockCode and summed the quantity sold for each item**
- **I converted the results to df to easy display it**
- **Finally I relpaced the index column that is generated by groupby (StockCode) with incremental index and added the StockCode column to the resulted df**

```python
zero_sold_items = grouped_items_sold[grouped_items_sold['QuantitySold'] <= 0]
zero_sold_items
```

```python
top_sold_items = grouped_items_sold[grouped_items_sold['QuantitySold'] > 0].sort_values(by='QuantitySold', ascending=False)
top_sold_items
```

```python
recommended_pairs = [
    {
        'top': top_sold_items.iloc[i]['StockCode'],
        'buttom': zero_sold_items.iloc[i]['StockCode']
    }
    for i in range(10)
]

recommended_pairs
```

<h2> The list above contains a dictionary of paired items stock_code's, one item from the top sold and the other from the lowest sold </h2>
